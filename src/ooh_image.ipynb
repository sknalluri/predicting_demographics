This notebook reads am xml file comntaining occupation titles and four image URLS to creates an HTML dataframe and webpage with four rendered images of each job description contained in the Bureau of Labor Statistics (BLS) Occupational Outlook Handbook (OOH). The images are those taken from thumbnails shown at the occupation group level (1) and from those shown in each of the following tabs at the individual occupation level (3): `What They Do`, `Work Environment`, `How to Become One`.

```python
import pandas as pd
import numpy as np
from IPython.core.display import HTML
```

```python
# Read images into df
ooh_images = pd.read_csv(r"OOH_Image_Data.txt", sep='|')
ooh_images
```

```python
from IPython.core.display import HTML
```

```python
# Function to convert links to html tags
def path_to_image_html(path):
    return '<img src="'+ path + '" width="60" >'
```

```python
# Rendering df as HTML table
ooh_images.to_html(escape=False, 
                   formatters=dict(Image_1=path_to_image_html,
                                   Image_2=path_to_image_html,
                                   Image_3=path_to_image_html,
                                   Image_4=path_to_image_html
                                  )
                  )
```

```python
# Rendering images in df using HTML method
HTML(ooh_images.to_html(escape=False,
                        formatters=dict(Image_1=path_to_image_html,
                                        Image_2=path_to_image_html,
                                        Image_3=path_to_image_html,
                                        Image_4=path_to_image_html
                                       )
                       )
    )
```

```python
# Saving the dataframe as a webpage
ooh_images.to_html("ooh_images.html",
                   escape=False, 
                   formatters=dict(Image_1=path_to_image_html,
                                   Image_2=path_to_image_html,
                                   Image_3=path_to_image_html,
                                   Image_4=path_to_image_html
                                  )
                  )
```
